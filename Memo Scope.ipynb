```python
def memo(f):
    """Decorator that caches the return value for each call to f(args). Then when called again with same args, we can just look it up"""
    
    cache={}
    
    def _f(*args):
        
        print ("cache is", cache)
        
        try:
            
            return cache[args]
        
        except KeyError:
            
            cache[args] = result = f(*args)
            
            return result
        
        except TypeError:
            # some element of args can't be a dict key
            
            return f(args)
        
    return _f

@memo
def fib(n):
    
    if n <= 1:
        
        return 1
    
    else:
        
        return fib(n-1) + fib(n-2)

@memo
def square(n):
    
    return n*n
            
print (fib(3))
print ("-------------------------------------------------")
print (fib(3))                #因為已經call過fib(3) => 所以在call一次fib(3)時，直接return cache裡的ans.(cache裡存有args = 3的ans.)，如圖1

print ("-------------------------------------------------")

print (square(3))             # scope不同 => 所以square(3)用的cache 與 fib(3)的cache是不同，如圖2 & 圖3
print ("-------------------------------------------------")
print (square(3))
```
